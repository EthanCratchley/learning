# SPY 10 Years

```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
# Read the CSV file
file_path = '/Users/ethancratchley/Desktop/SPY.csv'
df = pd.read_csv(file_path)

# Print the first few rows and get rid of index
print(df.head(10).to_string(index=False))
```

```python
# Convert 'Date' to datetime values
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

# Create the line chart
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Volume'], color='red', width=10)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('SPY 10 Years of Volume')

# Show the chart
plt.tight_layout()
plt.show()
```
